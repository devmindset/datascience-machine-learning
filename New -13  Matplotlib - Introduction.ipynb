```python
import matplotlib.pyplot as plt
#just to plot graph in ipython notebook
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
```

```python
plt.plot(x,np.sin(x), '.')
plt.show()
```

```python
plt.plot(x,np.cos(x),'-.')
```

```python
x = np.arange(10000)
y = np.square(x) 

plt.plot(x,y,'-')
plt.show()
```

### Subplotting

```python
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))

plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
```

```python
x = np.linspace(0,10,100)
plt.subplot(2,2,1)
plt.plot(x,np.sin(x))

plt.subplot(2,2,2)
plt.plot(x,np.cos(x))

plt.subplot(2,2,3)
plt.plot(x,np.cos(x))

plt.subplot(2,2,4)
plt.plot(x,np.cos(x))
```

```python
plt.subplot(1,2,1)
plt.plot(x,np.sin(x))

plt.subplot(1,2,2)
plt.plot(x,np.cos(x))
```

```python
fig, ax =  plt.subplots(3,2)
ax[0,0].plot(x,np.sin(x))
ax[0,1].plot(x,np.sin(x))
fig.show()
```
