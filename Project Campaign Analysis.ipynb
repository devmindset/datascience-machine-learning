```python
import numpy as np
```

```python
countries = selected_campaign_data['SE Country'].unique().tolist()
```

```python
countryleads = []
for country in countries:
    leads = np.sum(selected_campaign_data[selected_campaign_data['SE Country'] == country]['Total Leads'])
    countryleads.append((country,leads))
```

```python
top_countries_leads = sorted(countryleads,key= lambda d: d[1], reverse=True)[:10]
```

```python
marcoms = selected_campaign_data['Marcom Type'].unique().tolist()
marcomleads = []
for marcom in marcoms:
    leads = np.sum(selected_campaign_data[selected_campaign_data['Marcom Type'] == marcom]['Total Leads'])
    marcomleads.append((marcom,leads))
```

```python
sorted(marcomleads,key= lambda d: d[1], reverse=True)[:10]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import GradientBoostingRegressor

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
def create_pipeline(degree,estimator):
    return make_pipeline(PolynomialFeatures(degree=degree),estimator)
```

```python
p = PolynomialFeatures(3)
p.fit_transform([[2,3],[3,4]])
```

```python
models = [GradientBoostingRegressor(n_estimators=1000, max_depth=1, learning_rate=1.0),LinearRegression(fit_intercept=True)]
models_name= ['GBR','LR']
```

```python
X = featured_campaign_data[['SE Country','Marcom Type','Media Type','Campaign Level','Business Unit','Division','Quarter']]
y = featured_campaign_data['Total Leads']
for idx,model in enumerate(models):
    for degree in range(1,3):
        pipeline = create_pipeline(degree,model)
        pipeline.fit(X, y)
        y_pred = pipeline.predict(X)
        print (models_name[idx],degree,r2_score(y,y_pred))
```
