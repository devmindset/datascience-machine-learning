### Popular visualization libraries

* seaborn - lib built on top of matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
X = np.linspace(0,1,1000)
```

```python
Y = np.square(X) + 5*X + 7
```

```python
plt.plot(X,Y,'-')
plt.show()
```

```python
plt.plot(X, np.sin(X),'-k')
plt.plot(X, np.tan(X), '-r')
```

```python
X = np.arange(100)
```

```python
plt.plot(X,np.tan(X))
```

```python
plt.plot(np.arange(10),np.sinh(np.arange(10)))
```
