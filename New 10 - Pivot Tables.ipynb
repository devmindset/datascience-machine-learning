### Pivot Tables
* Finding relationship amon data similar to groupby
* Pivot tables are mutidimesional version of GroupBy

```python
import numpy as np
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.groupby('sex')[['survived']].mean()
```

```python
titanic.groupby(['sex','class'])['survived'].aggregate('mean').unstack()
```

```python
titanic.pivot_table('survived',index='sex',columns='class')
```

```python
help(titanic.pivot_table)
```

### Multi-level pivot table

```python
age = pd.cut(titanic['age'], [0,18,80])
```

```python
age
```

```python
titanic['age']
```

```python
titanic.pivot_table('survived', ['sex',age], 'class')
```

```python
titanic.pivot_table(index='sex', columns='class', aggfunc={'survived':sum, 'fare':np.mean})
```
