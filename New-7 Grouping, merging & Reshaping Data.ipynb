```python
import pandas as pd
goal_data = pd.read_csv('goal_stats_euro_leagues_2012-13.csv')
```

```python
goal_data.head()
```

```python
goal_data.set_index(['Month'],inplace=True)
```

```python
goalsScoredDF = goal_data.loc[goal_data['Stat']=='GoalsScored']
```

```python
goalsScoredDF.head()
```

```python
goalsScoredDF = goalsScoredDF.iloc[:,1:]
```

```python
goalsScored_grp = goalsScoredDF.groupby(lambda Month: Month.split('/')[2])
```

```python
#goalsScored_grp.mean()
goalsScored_grp.describe()
```

```python
goalsScored_grp.count()
goalsScored_grp.sum()
#help(goalsScored_grp.aggregate)
```

```python
for name,data in goalsScored_grp:
    print (name)
    print (data)
```

```python
fill_func = lambda x: x.fillna(x.mean())
```

```python
#Fill missing na in grouped by object
trans = goalsScored_grp.transform(fill_func)
```

```python
trans
```
