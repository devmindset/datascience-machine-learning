```python
import pandas as pd
```

```python
blog_data = pd.read_csv('data/blogData_train.csv',header=None)
```

```python
blog_data.info()
```

```python
blog_data.head()
```

```python
feature_col = [ col for col in blog_data.columns if col != 280 ]
```

```python
feature_data = blog_data[feature_col]
```

```python
target_data = blog_data[280]
```

```python
#finding how imp is other col with respect to target col
feature_corr = blog_data.corr().loc[280]
```

```python
feature_corr
```

```python
feature_df = pd.DataFrame({'col':feature_corr.index, 'imp':feature_corr.values})
```

```python
import numpy as np
feature_df['imp'] = np.abs(feature_df['imp'])
```

```python
feature_df
```

```python
#Taking top 50 features based on correlation
sec = feature_df.sort_values('imp',ascending=False)[:50]
```

```python
sec
```

```python
sec_cols = sec.col.values.tolist()
```

```python
#selecting data from top 50 columns
top50f = blog_data[sec_cols]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(top50f,target_data)
```

```python
test_blog_data = pd.read_csv('data/blogData_test-2012.02.04.00_00.csv',header=None)
```

```python
test_data = test_blog_data[sec_cols]
```

```python
test_target_data = test_blog_data[280]
```

```python
y_pred = model.predict(test_data)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(test_target_data,y_pred)
```
